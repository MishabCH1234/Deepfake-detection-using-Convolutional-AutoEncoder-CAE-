```python
import cv2
import os
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def crop_faces(image, boxes):
            faces = []
            for box in boxes:
                x1, y1, x2, y2 = box.astype(int)
                faces.append(image.crop((x1, y1, x2, y2)))
            return faces

def extract_frames(i,video_path, output_dir, frame_rate=1):
    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Open the video file
    video = cv2.VideoCapture(video_path)

    current_frame = 0
    success, image = video.read()
    while success:
        try:
          image=Image.fromarray(image)
        except TypeError:
          break
        boxes, probs = mtcnn.detect(image)
        if boxes is not None:
          image = crop_faces(image, boxes)[0]
          frame_filename = os.path.join(output_dir, f"frame_{i}_{current_frame:04d}.jpg")
          cv2.imwrite(frame_filename, np.array(image))

          # Skip to the next frame
          current_frame += frame_rate
          video.set(cv2.CAP_PROP_POS_FRAMES, current_frame)
          success, image = video.read()
        else:
          current_frame += frame_rate

    # Release the video capture object
    video.release()
    print(f"Extracted frames are saved in {output_dir} __ {i}")

mtcnn = MTCNN()
dirs=[os.path.join('/content/drive/MyDrive/New folder/Celeb-synthesis',i) for i in os.listdir('/content/drive/MyDrive/New folder/Celeb-synthesis')]
frame_rate = 50
for i, dir in enumerate(dirs):
  video_path = dir
  output_dir = '/content/drive/MyDrive/deep fake detection/Dataset/Train/Celeb-fake-frame'

  extract_frames(i,video_path, output_dir, frame_rate)
```
